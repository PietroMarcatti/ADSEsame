```python
import pandas as pd
import numpy as np
import itertools

n1 = pd.read_csv('../datasets/n1.csv')
n2 = pd.read_csv('../datasets/n2.csv')
n3 = pd.read_csv('../datasets/n3.csv')
n4 = pd.read_csv('../datasets/n4.csv')

all_paths = pd.concat([n1,n2,n3,n4], ignore_index= True)

def iou(vec1, vec2):
    intersection = np.minimum(vec1, vec2).sum()
    union = np.maximum(vec1, vec2).sum()
    return intersection / union if union != 0 else 0  # Avoid division by zero

iou_results = [0 for _ in range(195)]
# Step 2: Apply to each group
def compute_iou_for_group(group):
    node_vectors = group.iloc[:, 3:13].values  # Extract columns 3 to 13
    edge_vectors = group.iloc[:, 14:23].values
    num_rows = len(group)
    
    iou_sum_nodes = 0
    iou_sum_edges = 0
    count = 0
    
    for i in range(num_rows):
        for j in range(i+1,num_rows):
            iou_sum_nodes += iou(node_vectors[i], node_vectors[j])
            iou_sum_edges += iou(edge_vectors[i], edge_vectors[j])
            count+=1

    print(f"Finito il gruppo di {group.iloc[0,1]}")
    return {"node": iou_sum_nodes, "edge": iou_sum_edges, "count":count}

# Step 3: Group by 'Group' column and compute IoU
iou_results = all_paths.groupby('to').apply(compute_iou_for_group).reset_index(drop=True)
```

```python
import csv
csv_file = "../datasets/similarity.csv"
with open(csv_file, mode="w", newline="") as file:
    # Create a CSV DictWriter object
    writer = csv.DictWriter(file, fieldnames=["node", "edge", "count"])

    # Write the header (column names)
    writer.writeheader()

    # Write the rows from the list
    writer.writerows(iou_results)
```
